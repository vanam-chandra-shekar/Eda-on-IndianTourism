## outlier analysis

```python
plt.figure(figsize=(20, 6))
newdf.boxplot()
plt.show()
```

```python
def remove_outliers_zscore(df, columns, threshold=3):
    df_no_outliers = df.copy()
    
    for column in columns:
        if df_no_outliers[column].dtype in ['float64', 'int64']:
            z_scores = stats.zscore(df_no_outliers[column])
            abs_z_scores = abs(z_scores)
            df_no_outliers = df_no_outliers[abs_z_scores < threshold]
    
    return df_no_outliers
# Columns to check for outliers
numeric_columns = [
    'Rating',
    'Domestic visitor count 2020',
    'Domestic visitor count 2021',
    'Domestic visitor count 2022',
    'Foreign visitor count 2020',
    'Foreign visitor count 2021',
    'Foreign visitor count 2022',
]
# Remove outliers
df_cleaned = remove_outliers_zscore(newdf, numeric_columns)
```

```python
df_cleaned.shape
```

```python
# sanity checks
min_visitors, max_visitors = 10000, 5000000

# Apply the filter for each year and visitor type
df_cleaned = df_cleaned[
    (df_cleaned['Domestic visitor count 2020'].between(min_visitors, max_visitors)) &
    (df_cleaned['Domestic visitor count 2021'].between(min_visitors, max_visitors)) &
    (df_cleaned['Domestic visitor count 2022'].between(min_visitors, max_visitors)) &
    (df_cleaned['Foreign visitor count 2020'].between(min_visitors, max_visitors)) & 
    (df_cleaned['Foreign visitor count 2021'].between(min_visitors, max_visitors)) &
    (df_cleaned['Foreign visitor count 2022'].between(min_visitors, max_visitors)) 
]
#removing negetive visitors and visitors more than 5 lakhs
```

# Feature Engineering

```python
# making  a total Domestic visitors which makes it easy to get a idea of measure

df_cleaned["Total_Domestic_visitors"] = df_cleaned["Domestic visitor count 2020"] + df_cleaned["Domestic visitor count 2021"]+df_cleaned["Domestic visitor count 2022"]
df_cleaned["Total_Foreign_visitors"] = df_cleaned["Foreign visitor count 2020"] + df_cleaned["Foreign visitor count 2021"]+df_cleaned["Foreign visitor count 2022"]
```

```python
# creating featur of total visitor count
df_cleaned["Total Visitor Count"] = df_cleaned["Total_Domestic_visitors"] + df_cleaned["Total_Domestic_visitors"]
```

```python
#creating visitor ratio to capture the essential relation btween Domestic isitors and Foreign visitors

df_cleaned["Visitor Ratio"] = df_cleaned["Total_Domestic_visitors"] / df_cleaned["Total_Foreign_visitors"]
```

```python
# visulizing these features

fig , axes = plt.subplots(2 , 2, figsize=(10,8))

sns.histplot(df_cleaned["Total_Domestic_visitors"] , kde=True , ax=axes[0,0])
axes[0,0].set_title("Total_Domestic_visitors")

sns.histplot(df_cleaned["Total_Foreign_visitors"] , kde=True , ax=axes[0,1])
axes[0,1].set_title("Total_Foreign_visitors")

sns.histplot(df_cleaned["Total Visitor Count"] , kde=True , ax=axes[1,0])
axes[1,0].set_title("Total Visitor Count")

sns.histplot(df_cleaned["Visitor Ratio"] , kde=True , ax=axes[1,1])
axes[1,1].set_title("Visitor Ratio")

plt.tight_layout()
plt.show()
```

# Feature Scaling

```python
# As the values of the above columens ar astronomecialy high we heed to perfrom feature scaling 
# So that it does not effect the outpu tof the modal so much
# and also to get them as close as possible to normal distribution
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar = MinMaxScaler()
```

```python
df_cleaned["Total_Domestic_visitors"] = scalar.fit_transform(df_cleaned[["Total_Domestic_visitors"]])
df_cleaned["Total_Foreign_visitors"] = scalar.fit_transform(df_cleaned[["Total_Foreign_visitors"]])
df_cleaned["Total Visitor Count"] = scalar.fit_transform(df_cleaned[["Total Visitor Count"]])
df_cleaned["Visitor Ratio"] = scalar.fit_transform(df_cleaned[["Visitor Ratio"]])
```

```python
# visulizing these features

fig , axes = plt.subplots(2 , 2, figsize=(10,8))

sns.histplot(df_cleaned["Total_Domestic_visitors"] , kde=True , ax=axes[0,0])
axes[0,0].set_title("Total_Domestic_visitors")

sns.histplot(df_cleaned["Total_Foreign_visitors"] , kde=True , ax=axes[0,1])
axes[0,1].set_title("Total_Foreign_visitors")

sns.histplot(df_cleaned["Total Visitor Count"] , kde=True , ax=axes[1,0])
axes[1,0].set_title("Total Visitor Count")

sns.histplot(df_cleaned["Visitor Ratio"] , kde=True , ax=axes[1,1])
axes[1,1].set_title("Visitor Ratio")

plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA

# Select numerical columns for PCA
numerical_data = df_cleaned[['Rating', 'Total_Domestic_visitors', 'Total_Foreign_visitors', 'Total Visitor Count', 'Visitor Ratio']]

# Applying PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(numerical_data)

# Create a DataFrame with the principal components
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

# Scatter plot of the PCA results
plt.figure(figsize=(10, 6))
plt.scatter(pca_df['PC1'], pca_df['PC2'], marker='X', color='red')
plt.title('PCA of Visitor Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid()
plt.show()
```
